```python
# Given link
given_link = 'https://static.zara.net/photos///2024/V/0/1/p/2910/009/051/17/w/2048/2910009051_1_1_1.jpg?ts=1710424540258'

def extract3angleimages(link,dictio):
    # Detect key from the dictionary
    detected_key = None
    for key, value in dictio.items():
        if link == value:
            detected_key = key 
            break
    
    # Extract the 3 images from the same folder
    extracted_images = []
    for i in range(1,4):
        extracted_images.append(dictio[detected_key.split("/")[0] + "/" + str(i) + ".jpg"])
    return extracted_images
```
